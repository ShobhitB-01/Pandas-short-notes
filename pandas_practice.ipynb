```python
df['New Column'] = 100

df['New Column']
```

#### (v) Updating a single value

```python
df.head()
```

```python
df.loc[1, 'New Column'] = 69

df['New Column']
```

```python
df['Churn Binary'] = df['Churn'].apply(lambda x: 1 if x==True else 0)
```
